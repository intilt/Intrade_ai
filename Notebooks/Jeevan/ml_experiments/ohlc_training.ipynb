```python
def clean_dataset(df):
    assert isinstance(df, pd.DataFrame), "df needs to be a pd.DataFrame"
    df.dropna(inplace=True)
    indices_to_keep = ~df.isin([np.nan, np.inf, -np.inf]).any(axis=1)
    return df[indices_to_keep].astype(np.float64)
```

```python
def create_target(data, look_forward=50, threshold=0.008):
    # Compute forward returns
    data['return_fwd'] = np.log(data['close'].shift(-look_forward)) - np.log(data['close'])
    
    # Create binary target variable based on threshold
    data['target'] = np.where(data['return_fwd'] > threshold, 1, np.where(data['return_fwd'] < -threshold, 0, -1))
    
    # Drop missing values
    data.dropna(inplace=True)
    
    # Select relevant columns
    target = ['target']
    # threshold, buy=1, sell=0, hold=-1
    return data[target]
```

```python
def create_features(data):
    """Create features for daily OHLC data."""
    # Create lagged returns
    data['return_1'] = data['close'].pct_change(periods=1)
    data['return_5'] = data['close'].pct_change(periods=5)
    data['return_10'] = data['close'].pct_change(periods=10)
    
    # Create lagged volatility
    data['volatility_5'] = data['return_1'].rolling(window=5).std().shift(1)
    data['volatility_10'] = data['return_1'].rolling(window=10).std().shift(1)
    data['volatility_20'] = data['return_1'].rolling(window=20).std().shift(1)
    
    # # Create lagged volume
    # data['volume_1'] = data['volume'].pct_change(periods=1)
    # data['volume_5'] = data['volume'].pct_change(periods=5)
    # data['volume_10'] = data['volume'].pct_change(periods=10)
    
    # Create lagged momentum indicators
    data['sma_5'] = data['close'].rolling(window=5).mean().shift(1)
    data['sma_10'] = data['close'].rolling(window=10).mean().shift(1)
    data['sma_ratio'] = data['sma_5'] / data['sma_10']
    
    # Create binary variables for the day of the week
    data['day_of_week'] = data.index.dayofweek
    data['monday'] = np.where(data['day_of_week']==0, 1, 0)
    data['tuesday'] = np.where(data['day_of_week']==1, 1, 0)
    data['wednesday'] = np.where(data['day_of_week']==2, 1, 0)
    data['thursday'] = np.where(data['day_of_week']==3, 1, 0)
    data['friday'] = np.where(data['day_of_week']==4, 1, 0)
    
    # Create binary variables for the month of the year
    data['month'] = data.index.month
    data = pd.get_dummies(data, columns=['month'])

    data['target'] = create_target(data)
    
    # Drop missing values
    data.dropna(inplace=True)
    
    # Select relevant features
    features = ['return_1', 'return_5', 'return_10', 'volatility_5', 'volatility_10', 'volatility_20', 'sma_ratio', 'monday', 'tuesday', 'wednesday', 'thursday', 'friday', 'month_1', 'month_2', 'month_3', 'month_4', 'month_5', 'month_6', 'month_7', 'month_8', 'month_9', 'month_10', 'month_11', 'month_12','target']
    
    return clean_dataset(data[features])
```

```python
# data = nifty_1day.copy()
data = nifty_1min.copy()
df = data.copy()
data = create_features(data)
feature_names = data.columns.tolist()[:-1]

X = data.iloc[:, 0:-1]
y = data.iloc[:, -1]


# Split data into training and test sets
train_data = data.loc[data.index < '2020-05-01']
test_data = data.loc[data.index >= '2020-05-01']

# Create features for the training and test sets
X_train = train_data.iloc[:, 0:-1]
X_test = test_data.iloc[:, 0:-1]

# Create target variable for the training and test sets
y_train = train_data.iloc[:, -1]
y_test = test_data.iloc[:, -1]

# Normalize the feature data
scaler = StandardScaler()
X_train_norm = scaler.fit_transform(X_train)
X_test_norm = scaler.transform(X_test)

```
